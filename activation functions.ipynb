```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5, 5, 0.1)
print(x[1:10])
[-4.9 -4.8 -4.7 -4.6 -4.5 -4.4 -4.3 -4.2 -4.1]
```

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```python
y = [sigmoid(i) for i in x]
 
plt.axvline(x=0, color="red", linewidth=.5)
plt.axhline(y=0.5, color="red", linewidth=.5)
plt.plot(x, y)
plt.show()
```

```python
def tanh(x):
 return np.tanh(x)
```

```python
y = [tanh(i) for i in x]
 
plt.axvline(x=0, color="red", linewidth=.5)
plt.axhline(y=0, color="red", linewidth=.5)
plt.plot(x, y)
plt.show()
```

```python
def relu(x):
 return np.maximum(0, x)
```

```python
y = [relu(i) for i in x]
plt.axvline(x=0, color="red", linewidth=.5)
plt.axhline(y=0, color="red", linewidth=.5)
plt.plot(x, y)
plt.show()
```
